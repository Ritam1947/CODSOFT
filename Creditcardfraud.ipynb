Oversampling using SMOTE

```python
#ADVANTAGE OF USING SMOTE IS THAT WE ARE GENERATING SYNTHETIC DATAPOINTS THAT ARE SLIGHTLY DIFFERRENT FROM THE ORIGINAL DATAPOINTS
# AND NOT DUPLICATES DATAPOINTS
```

```python
x=df.drop('Class',axis=1)               #dropping class column
y=df['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res = SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
sns.barplot(x='Models', y='ACC', data=final_data)       #after oversampling as you can see random forest classifier is better than all the other models used
```
